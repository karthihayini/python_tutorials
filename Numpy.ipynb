```python
import datetime
```

```python
x = list(range(100000000))
y = list(range(100000000))
start = datetime.datetime.now()
# len(x + y)
z = [x[i]-y[i] for i in range(100000000)]
end = datetime.datetime.now()
time_taken_python = (end - start).microseconds
```

```python
x = np.arange(100000000)
y = np.arange(100000000)
start = datetime.datetime.now()
z = x - y
end = datetime.datetime.now()
time_taken_numpy = (end - start).microseconds
time_taken_numpy
```

```python
time_taken_numpy / time_taken_python
```

### Sorting

```python
np.random.seed(0)
x = np.random.randint(1,10,5) # Randomly select 5 elements between 1 to 10
x
```

```python
np.sort(x) # Ascending sorting
```

```python
np.sort(x)[::-1] # Descending sorting; [::-1] is used for reversing a list
```

```python
x
```

```python
np.argmax(x) # Index in which x has maximum value
```

```python
np.argmin(x) # Index in which x has minimum value
```

```python
x
```

```python
x.argsort()
```

```python
np.max(x)
```

## Stacking functions

```python
x1 = np.arange(10)
y1 = np.arange(10,20,1)
print(x1)
print(y1)
```

```python
np.hstack([x1, y1])
```

```python
np.vstack([x1, y1])
```
